# **Find all isolated pages**

```
#Count the number of dead-end pages
def dead_end(graph):
  dead_page_count = 0
  dead_page = []
  for key in graph.keys():
    if graph[key] == set():
      dead_page_count +=1
      dead_page.append(key)
  return dead_page_count, dead_page

#Count the number of isolated pages (hidden pages that cannot be visited from any other pages)
def isolated(graph, dead_page):
  isolated_page_count = 0
  isolated_page = []
  for page in dead_page:
    found = 0
    for key in graph.keys():
      if page in graph[key]:
        found = 1
        break
    if found == 0:
      isolated_page_count +=1
      isolated_page.append(page)
  return isolated_page_count, isolated_page


dead_page_count, dead_page = dead_end(graph)
isolated_page_count, isolated_page = isolated(graph, dead_page)

print('=================================')
print('The number of total pages:')
print(len(dic_page))
print('=================================')
print('The number of dead-end pages (No pointers to other pages):')
print(dead_page_count)
print('=================================')
print('The number of isolated pages (hidden pages that cannot be visited from any other pages:')
print(isolated_page_count)
print('=================================')
```

# **Find the shortest path from Google to Shibuya?**

```
from collections import deque

#Find shortest path from start_node to end_node
def bfs(visited_node, graph, start_node, end_node):
  #store all the paths
  queue = deque([[start_node]])
  while len(queue) > 0:
    current_path = queue.popleft()
    current_node = current_path[-1]
    if current_node not in visited_node:
      neighbours = graph[current_node]
      for neighbour in neighbours:
        #form new path
        temp = current_path.copy()
        temp.append(neighbour)
        if neighbour == end_node:
          return temp
        queue.append(temp)
  return None
    

page_1 = 'Google'
page_2 = '渋谷'
visited_node = []
visited_node = bfs(visited_node, graph, page_1, page_2)
print(visited_node)
```

# **Farthest page from Google**

```
#Find the farthest page from Google
from collections import deque
def bfs_distance(visited_node, graph, start_node):
  #A dictionary to store distance for each page from Google
  distance = {}
  distance[start_node] = 0
  queue = deque()
  queue.append(start_node)
  while len(queue) > 0:
    current_node = queue.popleft() 
    for neighbour in graph[current_node]:
      if neighbour not in visited_node:
        visited_node.add(neighbour)
        queue.append(neighbour)
        #add 1 to distance for current 'neighbour'
        distance[neighbour] = distance[current_node]+1
  return distance

visited_node = set()
visited_node.add('Google')
distance = bfs_distance(visited_node, graph, 'Google')  
distance = sorted(distance.items(), key=lambda d: d[1]) 
max_distance = [pair for pair in distance if pair[1]==distance[-1][1]]
print('======================================')
print('Max distance from Google:')
print(max_distance[0][1])
print('======================================')
print('Page name:')
for page in max_distance:
  print(page[0])
```
